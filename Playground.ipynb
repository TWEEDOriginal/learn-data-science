```python
# define the funciton
def f(x, mu, s2):
    return 1/np.sqrt(2*np.pi*s2)*np.exp(-(x-mu)**2/2/s2)

x = np.linspace(-5, 5, 1000)

plt.plot(x, f(x, 0, 0.2), 'b', label="$\mu=0, \sigma^2=0.2$")
plt.plot(x, f(x, 0, 1.0), 'r', label="$\mu=0, \sigma^2=1.0$")
plt.plot(x, f(x, 0, 5.0), 'orange', label="$\mu=0, \sigma^2=5.0$")
plt.plot(x, f(x, -2, 0.5), 'g', label="$\mu=-2, \sigma^2=0.5$")
plt.grid()

plt.xlabel('$X$')
plt.ylabel('$\phi_{\mu, \sigma^2}(X)$')
plt.yticks(np.arange(0, 1.2, 0.2))
plt.xticks(np.arange(-5, 6, 1))

plt.show()
```
